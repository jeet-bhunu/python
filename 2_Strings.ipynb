## <span style="color: #006400">String Properties in Python</span>
<br>
1. <span style="color: #800000;font-weight:bolder;font-size:18px">Immutable</span><br>
Once you make a string, you can't change it! It's like you've made a necklace of beads and glued them together. You can't take a bead off or add one on, but you can make a whole new necklace with some different beads.<br>
<br>
2. <span style="color: #800000;font-weight:bolder;font-size:18px">Sequences</span><br>
Just like each bead on a necklace has its own place in the sequence, each character in a string has its own index. You can use these indices to access individual characters in a string. For example, to get the first character of a string <span style="color: #0000FF">my_string</span>, you would write <span style="color: #0000FF">my_string[0]</span>. And just like you can count the number of beads on a necklace, you can count the number of characters in a string using the <span style="color: #000080">len()</span> function.<br>
<br>
3. <span style="color: #800000;font-weight:bolder;font-size:18px">Operations</span><br>
You can do lots of cool things with strings! You can add them together to make a new string, or even multiply them by a number to repeat them. You can also use various string methods to manipulate them, like converting them to uppercase or lowercase, replacing parts of them with other strings, or splitting them into a list of words.

```python
my_name='hello,jeet'

print(my_name)


my_Name='jeet'
last_name='bhunu'
full_name=my_name+" "+last_name
print(full_name)
```

```python
" s ".join([my_Name, last_name])
```

## <p style="color:#008CBA">&#x1F4E2; Accessing and Manipulating Strings &#x1F4E2;</p>

<p style="color:#777">Now that we know how to create strings, let's learn how to access and manipulate them. &#x1F9D0;</p>

<ol>
  <li>
    <p>We can access individual characters in a string using indexing, like this:</p>
    <pre><code>my_string = "Hello, world!"
print(my_string[0])   # Output: <span style="color:green;">"H" &#x1F44D;</span>
print(my_string[4])   # Output: <span style="color:green;">"o" &#x1F601;</span>
print(my_string[-1])  # Output: <span style="color:green;">"!" &#x1F60A;</span></code></pre>
  </li>
  <li>
    <p>We can also access a range of characters in a string using slicing, like this:</p>
    <pre><code>my_string = "Hello, world!"

print(my_string[7:])    # Output: <span style="color:green;">"world!" &#x1F389;</span></code></pre>
  </li>
  <li>
    <p>We can concatenate strings using the <code>+</code> operator, like this:</p>
    <pre><code>first_name = "John"

print(full_name)   # Output: <span style="color:green;">"John Doe" &#x1F60E;</span></code></pre>
  </li>
  <li>
    <p>We can also repeat a string multiple times using the <code>*</code> operator, like this:</p>

  </li>
  <li>
    <p>We can convert a string to uppercase or lowercase using the <code>upper()</code> and <code>lower()</code> methods, like this:</p>
    <pre><code>my_string = "Hello, World!"
print(my_string.upper())   # Output: <span style="color:green;">"HELLO, WORLD!" &#x1F4AA;</span>

# <p style="color:#008CBA">&#x1F4AC; String Methods</p>

<p style="color:#777">Imagine a symphony orchestra where each musician has their own unique instrument. Just like each musician, each string in a string data type has its own unique qualities and characteristics. And just like how a musician can use various techniques to manipulate the sound produced by their instrument, we can use various string methods to manipulate the strings in Python.</p>
<p>Let's take a look at some of the most commonly used string methods and how they relate to our orchestra analogy:</p>

```python
path=r"C:\Users\user\Desktop\mbox-short.txt"
file=open(path).read()
print(file)
```


<h3 style="color:#008CBA">&#x1F3BC; .capitalize()</h3>
<p>Think of the .capitalize() method as the conductor of the orchestra, setting the tone and starting the performance. This method capitalizes the first letter of a string, giving it a strong and commanding presence.</p>

```python
my_string="hello, jeet"

stings=my_string.capitalize()

print(stings)
```

<h3 style="color:#008CBA">&#x1F3BB; .casefold()</h3>
<p>The .casefold() method is like the orchestra's tuning session. Just as musicians tune their instruments to ensure they are playing in harmony, the .casefold() method ensures that all the letters in a string are in the same case, avoiding any inconsistencies that may cause confusion.</p>


```python
s=my_string.casefold()

print(s)
```

<h3 style="color:#008CBA">&#x1F3B9; .join()</h3>
<p>The .join() method is like a string quartet, where each musician plays their part to create a beautiful and cohesive sound. This method joins a sequence of strings together, creating a new string that combines all the unique qualities of each individual string.</p>

```python
s1='jeet'
s2='bhunu'
" s ".join([s1,s2])
```

<h3 style="color:#008CBA">&#x1F3A4; .find()</h3>
<p>The .find() method is like a detective, searching for a specific character or substring within a string. This method returns the index of the first occurrence of the specified character or substring, giving us a clue as to where we can find what we're looking for.</p>

```python
my_string.find('j')
```

<h3 style="color:#008CBA">&#x1F3A7; .replace()</h3>
<p>The .replace() method is like a musician swapping out their instrument for a new one. This method replaces a specified substring with another substring, allowing us to switch out specific parts of a string as needed.</p>

```python
my_string.replace('bye')
```

<h3 style="color:#008CBA">&#x1F3B8; .split()</h3>
<p>The .split() method is like a music stand, separating individual sheet music into manageable parts. This method splits a string into a list of substrings based on a specified separator, making it easier to access and work with each individual part of the string.</p>

<h3 style="color:#008CBA">&#x1F3A5; .isdigit()</h3>
<p>The .isdigit() method is like a ticket inspector, ensuring that each ticket is valid and numeric. This method checks whether all the characters in a string are digits, helping us determine whether the string can be used for numeric operations.</p>

<h3 style="color:#008CBA">&#x1F3B5; .strip()</h3>
<p>The .strip() method is like a wardrobe assistant, ensuring that each musician is dressed appropriately for their performance. This method removes any leading or trailing whitespace from a string, ensuring that the string is clean and polished before it goes on stage.</p>


<h3 style="color:#008CBA">&#x1F3A8; .lower()</h3>
<p>The .lower() method is like a volume knob that turns down the volume of a singer who is singing too loudly. This method makes all the characters in a string lowercase, making it easier to work with when comparing or searching for specific text.</p>
<p>For example:</p>
<pre><code>my_string = "Hello World"
print(my_string.lower())   # Output: "hello world"</code></pre>

<h3 style="color:#008CBA">&#x1F3A8; .upper()</h3>
<p>The .upper() method is like a volume knob that turns up the volume of a singer who is too quiet. This method makes all the characters in a string uppercase, making it easier to work with when comparing or searching for specific text.</p>

<h3 style="color:#008CBA">&#x1F3A8; .title()</h3>
<p>The .title() method is like a spotlight that shines on the lead singer of a band. This method capitalizes the first letter of each word in a string, making it look like a title or headline.</p>
